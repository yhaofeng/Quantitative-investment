```python
from vine import vine
```

```python
v = vine(df)
```

```python
v.fit()
probs = v.conditional_probability()
```

```python
d = pd.DataFrame()
d['probs'] = probs
d
```

```python
d['A'] = df['中债国债到期收益率:2个月']
d['B'] = df['中债国债到期收益率:3个月']
d = d[['A','B','probs']]
d
```

```python
d['probs'] = d['probs']-0.5
d
```

```python
import gym
import fin_gym
```

```python
env = gym.make(id='fin_env-v0',data=values,gamma=0.9,dates=dates,action_shape=3,fee_A=0.01,fee_B=0.01)
```

```python
from finrl.drl_agents.stablebaselines3.models import DRLAgent
```

```python
agent = DRLAgent(env = env)
```

```python
A2C_PARAMS = {"n_steps": 5, "ent_coef": 0.005, "learning_rate": 0.0002}
model_a2c = agent.get_model(model_name="a2c",model_kwargs = A2C_PARAMS)
```

```python
trained_a2c = agent.train_model(model=model_a2c, 
                                tb_log_name='a2c',
                                total_timesteps=50000)
```
